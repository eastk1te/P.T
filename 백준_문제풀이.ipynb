#### 4673번 셀프넘버 : numpy 모듈 사용 불가

```python
def self_num(n):
  x = 0
  for i in range(len(str(n))):
    x += int(str(n)[i])
  x += n
  return x

def self_num_list(n):
  x = []
  while True:
    if self_num(n) > 10001:
      break
    x.append(self_num(n))
    n = self_num(n)    
  return x

c = 0
b = []

for i in range(1,10001):
  b.append(i)

while True:
  
  b = list(set(b) - set(self_num_list(b[c])))
  b.sort()
  c += 1
  if c > len(b)-1:
    break

for i in b:
  print(i)
```

#### 1065번 한수 

```python
def han(n):
  res = 0
  if n < 100:
    res = 1
  elif n >= 100:
    n = str(n)
    if int(n[2]) + int(n[0]) == 2 * int(n[1]):
      res = 1
    else:
      res = 0

  elif n >= 1000:
    if (int(n[2]) + int(n[0]) == 2 * int(n[1])) & (int(n[3]) + int(n[1]) == 2 * int(n[2])):
      res = 1
    else:
      res = 0

  else:
    pass

  return res

def count_han(n):
  cnt = 0
  for i in range(n):
    cnt += han(i+1)
  
  return cnt

a = int(input())
print(count_han(a))
```
