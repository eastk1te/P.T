## DataFrame 사용

```
accident = pd.DataFrame(data=data)
accident = accident.iloc[:,0:4]
accident['사망비율'] = accident.사망자수 / accident.사고건수
accident.dropna(how='all')
accident
```

## 데이터 연산 및 필터링

```
age_total = accident['사고건수'].groupby(accident['가해자연령층']).sum()
age_total
```

```
month_total = accident['사고건수'].groupby(accident['월']).sum()
month_total
```
