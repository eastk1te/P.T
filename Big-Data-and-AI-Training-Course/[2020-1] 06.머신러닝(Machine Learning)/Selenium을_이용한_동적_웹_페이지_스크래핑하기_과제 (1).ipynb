```
def scraping(played=13, page_num=1):
  wd = webdriver.Chrome('chromedriver',options=chrome_options)
  wd.implicitly_wait(3)

  query = quote(input('검색어 입력:'))
  search_url = 'https://www.op.gg/summoner/userName=' + query
  wd.get(search_url)
  print(search_url)

  user_idx = 0
  page_idx = 0
  list_idx = 2
  user_df = pd.DataFrame(columns=('Result','Champion','Kill','Death','Assist','Damage','CS','CSperMIN'))

  
  
  while True:
    try:
      button = wd.find_element_by_xpath(f'//*[@id="SummonerLayoutContent"]/div[2]/div[2]/div/div[2]/div[{list_idx+2}]/a')
      button.send_keys('\n')
      time.sleep(1)
      page_idx += 1
      list_idx += 1
      print('[더 보기]')

      if page_idx > page_num:
        pass

    except:
      print('[더 보기 없음]')
      pass

    try:
      game_lists = wd.find_elements_by_class_name('GameItemList')
      print('[게임결과]')
      list_idx = 2

      for game_list in game_lists:
        list_idx += 1
        games = game_list.find_elements_by_class_name('GameItemWrap')

        for game in games:
          result = game.find_element_by_class_name('GameResult').text
          champion = game.find_element_by_class_name('ChampionName').text
          kill = game.find_element_by_class_name('Kill').text
          death = game.find_element_by_class_name('Death').text
          assist = game.find_element_by_class_name('Assist').text
          
          btn = game.find_element_by_id('right_match')
          btn.send_keys('\n')
          time.sleep(1)

          detail = game.find_element_by_class_name('GameDetail')

          damage = detail.find_element_by_class_name('ChampionDamage').text
          cs = wd.find_element_by_xpath(f'//*[@id="SummonerLayoutContent"]/div[2]/div[2]/div/div[2]/div[{list_idx}]/div[{user_idx + 1}]/div/div[2]/div/div[2]/div[1]/div/table[1]/tbody/tr[1]/td[10]/div[1]').text
          cs_min = detail.find_element_by_class_name('CSPerMinute').text

          user_df.loc[user_idx] = [result, champion,kill,death,assist,damage,cs, cs_min]
          user_idx += 1
          print(f'최근 {user_idx} 번째 게임')
          print([result, champion,kill,death,assist,damage,cs, cs_min])
          if user_idx >= played:
            break
        if user_idx >= played:
          break
      if user_idx >= played:
        break
      
    except NoSuchElementException as e:
      print(e)
      pass

  wd.close()
  
  return user_df

```
