```
img = cv2.imread('/content/bee-5618012_640.jpg')
```

```
plt.imshow(img)
plt.show()
```

```
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

- Flip Upside Down and display it in the notebook

```
img = cv2.flip(img, -1)
plt.imshow(img)
plt.show()
```

- Draw an empty RED rectangle around dogs face and display it

```
img_1 = cv2.flip(img,-1)
img_1 = cv2.rectangle(img_1,(280,100),(420,230),(255,0,0),2)
plt.imshow(img_1)
plt.show()
```

- Draw a BLUE TRIANGLE in the middle of the image, and display it

```
img_2 = cv2.flip(img,-1)
pts = np.array([[300,150],[200,300],[400,300],[300,150]], np.int32)
img_2 = cv2.polylines(img_2,[pts],False,(0,0,255),3)
plt.imshow(img_2)
plt.show()
```

- fill in the triangle
  - `cv2.fillPoly()`

```
img_2 = cv2.flip(img,-1)
pts = np.array([[300,150],[200,300],[400,300],[300,150]], np.int32)
img_2 = cv2.fillPoly(img_2,[pts],(0,0,255),8,0)
plt.imshow(img_2)
plt.show()
```
