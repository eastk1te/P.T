- `cv2.create()`

      - `PANORAMA`

      - `SCANS`

- `cv2.stitch()`

  - `Parameters`

    - `images` : 입력 영상의 벡터

    - `pano` : 출력 파노라마 영상

  - `반환값` : 함수 동작 결과 코드


```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from google.colab import files
files.upload()
```

```
imgs = []
```

```
img1 = cv2.imread('/content/bee_1.jpg')
img2 = cv2.imread('/content/bee_2.jpg')
```

```
cv2_imshow(img1)
```

```
cv2_imshow(img2)
```

```
imgs.append(img1)
imgs.append(img2)
```

```
len(imgs)
```

```
stitcher = cv2.Stitcher_create()
```

```
status, dst = stitcher.stitch(imgs)
```

```
cv2_imshow(dst)
```
