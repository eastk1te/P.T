```
img = cv2.imread('/content/KakaoTalk_20201103_195238645.jpg',0)
```

- Apply a binary threshold onto the image.

```
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```
ret, thresh = cv2.threshold(img,128,255,cv2.THRESH_BINARY)
```

```
plt.figure(figsize=(6,8))
plt.imshow(thresh,cmap='gray')
plt.show()
```

- convert colorspace to HSV, and display it

```
img = cv2.imread('/content/KakaoTalk_20201103_195238645.jpg')
```

```
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
plt.show()
```

- Create a low pass filter with a 4 by 4 kernel filled with values of 1/10(0.1), and then use 2D convolution to blur the image 

```
kernel = np.ones((4,4),np.float32)/(10)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```
kernel
```

```
blur = cv2.filter2D(img, -1,kernel)
```

```
plt.imshow(blur)
plt.show()
```

- Create a Horizontal Sobel Filter with a kernel size of 5 to the grayscale version of the image, and display the result

```
img = cv2.imread('/content/KakaoTalk_20201103_195238645.jpg',0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
sobelx = cv2.Sobel(img, cv2.CV_8U, 1,0,ksize=5)
plt.imshow(sobelx, cmap='gray')
plt.show()
```

- Plot the color histograms for the RED, BLUE, and GREEN channel of the image.

```
img = cv2.imread('/content/KakaoTalk_20201103_195238645.jpg')
R = cv2.calcHist(img,[3],None,[256],[0,256])
G = cv2.calcHist(img,[2],None,[256],[0,256])
B = cv2.calcHist(img,[1],None,[256],[0,256])
```

```
plt.figure(figsize=(12,8))
plt.plot(R,color='r')
plt.plot(G,color='g')
plt.plot(B,color='')
plt.xlim([0,256])
plt.show()
```
