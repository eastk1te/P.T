## 입력값이 양수, 음수, 0인지 구분하는 코드를 작성하시오. (20점)

```
n = int(input("정수 입력:"))
if n > 0:
  print('입력한 정수는 : 양수')
elif n == 0:
  print('입력한 정수는 : 0')
else:
  print('입력한 정수는 : 음수')
#코드 작성
```

## 삼각형과 사각형의 넓이는 구하는 함수를 작성하시오. (20점)

```
def triangle(w, h):
  print('삼각형의 넓이: ', w*h*0.5)

def rectangle(w, h):
  print('사각형의 넓이: ', w*h)

width = int(input('너비 입력:'))
height = int(input('높이 입력:'))

triangle(width,height)
rectangle(width,height)
#코드 작성
#삼각형과 사각형 넓이 출력
```

## 두 수를 입력받아 계산하는 함수들을 작성하시오. (20점)

```
def add(a, b) :
	return a + b

def sub(a, b) :
	return a - b

def mult(a, b) :
	return a * b

def div(a, b) :
  return a // b
	#코드 입력

x = int(input('정수 x: '))
y = int(input('정수 y: '))

print('두 수의 덧셈 : %d' % add(x, y))
print('두 수의 뺄셈 : %d' % sub(x, y))
print('두 수의 곱셈 : %d' % mult(x, y))
print('두 수의 나눗셈 : %d' % div(x, y))
```

## 타자 게임 코드를 발전시켜 보세요. (20점)

```
import random
import time

# 단어 리스트 w 선언과 초기화!!
w = ['cat', 'bird','dog', 'fox', 'monkey', 'lion', 'tiger', 'mouse', 'panda', 'frog', 'snake', 'wolf', 'horse']

n = 1 # 문제 번호로 사용될 변수

print('[타자 게임] 준비되면 엔터를 누르세요.')
input()

start = time.time() # 엔터 누르고 나서의 현재 시간을 start에 대입!

q = random.choice(w) # 리스트 w 내의 값 중 하나가 선택
s = 0
while n <= 5 : # n이 5 이하이면 참
    print('** 문제', n, '번 **')
    print(q)
    x = input()
    if q == x :
        print('성공')
        n = n + 1
        s += 20
        print('점수: ',s)
        q = random.choice(w) # 리스트 w의 다른 내용으로 q에 대입
    else :
        print('실패')
        print('점수: ',s)
        n = n + 1
        q = random.choice(w)
        continue

end = time.time()
print('\n-------------------결과창------------------------')
if s <= 20:
  print('다시 한번 해보세요!')
elif s <= 40:
  print('조금 더 열심히!')
elif s <= 60:
  print('조금 아쉽네요')
elif s <= 80:
  print('잘했어요!')
else:
  print('매우 잘했어요!')


print('점수: ',s)
et = end - start
et = format(et, '.2f') # et의 값을 소수점 2자리 실수로 표현!

print('타자 시간 :', et, '초')
```
