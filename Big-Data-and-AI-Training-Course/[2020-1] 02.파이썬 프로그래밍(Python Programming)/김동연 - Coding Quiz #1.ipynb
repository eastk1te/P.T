## 1부터 100 사이의 홀수와 짝수의 누적합을 구하시오. (10점)

```
even_sum = 0
odd_sum = 0

for i in range(1, 101):
  if i % 2 == 0:
    even_sum = even_sum + i
  else:
    odd_sum += i

print(even_sum)
print(odd_sum)# 코드 입력
```

## 2부터 100 사이의 소수를 구하는 코드를 작성하시오. (30점)

```
for i in range(2, 101):
  if i == 2:
    list = [2]
  else:
    for j in range(2,i):
      if i % j == 0:
        break
      elif j == i - 1:
        list.extend([i])
print(list)
```

## 나이에 따른 입장료를 계산하는 코드를 작성하시오. (30점)

* 1~7세: 1000원
* 8~20세: 2000원
* 21~40세: 3000원
* 41~세: 1500원

```
age = 17
pay = 0

if age <= 7:
  pay = 1000
elif age <= 20:
  pay = 2000
elif age <= 40:
  pay = 3000
else:
  pay = 1500
# 코드 입력

print("나이: {0}, 입장료: {1}".format(age, pay))
```
