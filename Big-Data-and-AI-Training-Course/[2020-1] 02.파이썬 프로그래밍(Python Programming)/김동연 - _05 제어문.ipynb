## 조건문(Conditional Statement)


### if 문


* if 문을 이용한 미세먼지 측정
  + 미세먼지 농도 pm
  + 35 초과는 미세먼지 농도 나쁨

```
n =40
if n >35 :
  print('미세먼지 농도 나쁨')
```

### if-else 문


* if-else 문을 이용한 미세먼지 측정
  + 미세먼지 농도 pm
  + 36 이상은 미세먼지 농도 나쁨
  + 35 이하는 미세먼지 농도 좋음

```
n =30
if n >= 36 :
  print('미세먼지 농도 나쁨')
else:
  print('미세먼지 농도 좋음')
```

### if-elif-else 문


* if-elif-else 문을 이용한 미세먼지 측정
  + 미세먼지 농도 pm
  + 미세먼지 농도 0~15: 좋음
  + 미세먼지 농도 16~35: 보통
  + 미세먼지 농도 36~75: 나쁨
  + 미세먼지 농도 76~: 매우나쁨

```
n =100
if n <= 15 :
  print('미세먼지 농도 좋음')
elif n <= 35:
  print('미세먼지 농도 보통')
elif n <= 75:
  print('미세먼지 농도 나쁨')
else:
  print('미세먼지 농도 매우나쁨')
  
```

### 중첩 if 문




* 중첩 if 문을 이용한 미세먼지 측정
  + 미세먼지 농도 pm
  + 미세먼지 농도 0~15: 좋음
  + 미세먼지 농도 16~35: 보통
  + 미세먼지 농도 36~75: 나쁨
  + 미세먼지 농도 76~: 매우나쁨

```
n =30
if n <= 35 :
  if n <= 15:
   print('미세먼지 농도 좋음')
  else:
   print('미세먼지 농도 보통')    
else:
  if n <= 75:
    print('미세먼지 농도 나쁨')
  else :
    print('미세먼지 농도 매우나쁨')
```

### if-pass 문

* 조건문은 있지만 실행할 문장이 없는 경우, 오류가 발생하지 않도록 무시하고 넘어가는 기능

```
if 10 > 5:
  print(10)
else:
  pass

if 10 < 5:
  print(10)
else:
  pass
```

### if 조건 연산자

* 비교연산자: <, >, ==, !=, >=, <=


```
if 2>1:
  print(2)
if 3==3:
  print(3)
if 1 != 2:
  print(1)
```

* 논리연산자: and, or, not

```
rain = True
snow = True
sun = 1
if rain and snow:
  print('진눈깨비')

if not sun :
  print('흐림')
else:
  print('맑음')
```

* 멤버연산자: in, not in


```
list = ['one','two','three']

if 'one' in list:
  print('one')

if 'four' not in list:
  print('four')
```

* 식별연산자: is, is not

```
if 'one' is 'one':
  print('one')

if 'one' is not 'two':
  print('one is not two')
```

### 조건부 표현식(Conditional Expression)

* 한 라인으로 조건식을 사용한 표현

```
score = 75
msg = "통과" if score <= 70 else "탈락"
print(msg)
```

## 반복문(Repetitive Statement)


### while 문


* while 문 예제: 1부터 10까지 반복

```
i =1
while i <=10:
  print(i)
  i = i + 1 # i += 1
```

* while 문 예제: 1부터 10까지 더하기

```
i =1
sum = 0
while i <=10:
  sum = sum + i # sum += i
  i += 1

print(sum)
```

### for 문


* 리스트 요소 반복

```
list = ['one','two','three']
for i in list:
  print(i)
```

* 튜플 요소 반복

```
tuple = ('one','two','three')
for i in tuple:
  print(i)
```

* 문자열 요소 반복

```
string = 'dongyeonkim'
for i in string:
  print(i)
```

#### range()


```
sum = 0
for i in range(11):
  sum += i

print(sum)
```

### break 문

* 반복문 종료

```
i = 0
while i < 100:
  print(i)
  if i == 10:
    break
  i += 1
```

```
for i in range(100):
  print(i)
  if i ==10:
    break
```

### continue 문

* 반복 조건문으로 이동

```
i = 0
while i < 10:
  i += 1
  if i % 2== 0:
    continue
  print(i)
```

```
for i in range(10):
  if i % 2 ==0:
    continue
  print(i)
```

### 리스트 내포(List Comprehension)

* 리스트 안에 for 문과 if 문 사용

```
list = [1,2,3,4,5]
print([i*2 for i in list])
print([i*2 for i in range(10) if i % 2 == 0])
```

```
name + 1 
```

```
'1' + 1
```

```
int('string')
```

### 예외 처리(Exception Handling)

#### try, except 문

* try 문을 통해 예외 발생 가능한 코드에 정의
* except 문을 이용해 예외 처리 방법을 정의

```
try:
  print( 2/0)
except:
  print("error")
```

#### 에러 분류에 따른 except 문

* `ZeroDivisionError`에 대한 except 처리

```
try:
  print(2/0)
except ZeroDivisionError as e :
  print('error')
```

#### 에러 메시지가 포함된 except 문

* `ZeroDivisionError`에 포함된 에러 메시지 `e`를 출력

```
try:
  print(2/0)
except ZeroDivisionError as e :
  print(e)
```

#### 여러 에러 분류가 포함된 except 문

```
try:
  n = int(input('입력: '))
  result = 1 / n
except ZeroDivisionError as e:
  print(e)
except ValueError as e:
  print(e)
else:
  print("결과 : " + str(result))
```

### 에러 발생

#### raise 문

* 직접적으로 에러를 발생시킬 수 있음
* 에러의 종류와 에러 상황을 정의하고 발생

```
raise ZeroDivisionError("숫자는 0으로 나눔")
```

```
raise NameError("지역 또는 전역 이름을 찾을 수 없음")
```

```
raise TypeError("연산이나 함수가부적절한 형의 객체에 적용")
```

```
raise ValueError("연산이나 함수가 부적절한 값을 인자로 받음")
```

* 검증식이 `False`일 경우 `AssertionErro` 발생

```
assert True
```

```
assert False
```
