```
import tensorflow as tf
from tensorflow.keras.datasets.boston_housing import load_data
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, ReLU, BatchNormalization,Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l1, l2, l1_l2

from sklearn.model_selection import train_test_split
import numpy as np
```

```
tf.random.set_seed(999)

(x_train_full, y_train_full), (x_test, y_test) = load_data(path='boston_housing.npz',
                                                           test_split=0.3,
                                                           seed=999)
```

```
mean = np.mean(x_train_full, axis=0)
std = np.std(x_train_full, axis=0)

x_train_preprocessed = (x_train_full - mean) / std
x_test = (x_test - mean) / std

x_train, x_val, y_train, y_val = train_test_split(x_train_preprocessed, y_train_full,
                                                  test_size=0.3,
                                                  random_state=111)
```

### 모델1
 - model1.evaluate(x_test,y_test)
 - [28.100740432739258, 3.2786977291107178]

```
model1 = Sequential([Dense(100, input_shape=(13,), name='dense1'),
                    Dense(64, activation='relu', name='dense2'),
                    Dense(32, activation='relu', name='dense3'),
                    Dense(1, name='output'),])
```

```
model1.compile(loss='mse',
              optimizer =Adam(learning_rate=1e-2),
              metrics=['mae'])
```

```
model1.fit(x_train, y_train, epochs=300,
           validation_data=(x_val, y_val))
```

```
model1.evaluate(x_test,y_test)
```

### 모델2
 - 차원축소로 성능 향상

```
model2 = Sequential([Dense(64, activation='relu', input_shape=(13,), name='dense1'),
                    Dense(64, activation='relu', name='dense2'),
                    Dense(32, activation='relu', name='dense3'),
                    Dense(1, name='output'),])
```

```
model2.compile(loss='mse',
              optimizer =Adam(learning_rate=1e-2),
              metrics=['mae'])
```

```
model2.fit(x_train, y_train, epochs=300,
           validation_data=(x_val, y_val))
```

```
model2.evaluate(x_test,y_test)
```

### 모델3
 - 가중치 초기화 및 배치정규화로 성능 향상

```
model3 = Sequential([Dense(64,kernel_initializer='he_normal', input_shape=(13,), name='dense1'),
                     BatchNormalization(),
                     ReLU(),
                     Dense(64,kernel_initializer='he_normal', name='dense2'),
                     BatchNormalization(),
                     ReLU(),
                     Dense(32,kernel_initializer='he_normal', name='dense3'),
                     BatchNormalization(),
                     ReLU(),
                     Dense(1,kernel_initializer='he_normal', name='output')])
```

```
model3.compile(loss='mse',
              optimizer =Adam(learning_rate=1e-2),
              metrics=['mae'])
```

```
model3.fit(x_train, y_train, epochs=300,
           validation_data=(x_val, y_val))
```

```
model3.evaluate(x_test,y_test)
```

### 모델4
 - 규제화를 통해 성능향상
 - 가장 높은 성능

```
model4 = Sequential([Dense(64,kernel_initializer='he_normal',kernel_regularizer=l2(0.001), input_shape=(13,), name='dense1'),
                     BatchNormalization(),
                     ReLU(),
                     Dense(64,kernel_initializer='he_normal',kernel_regularizer=l2(0.001), name='dense2'),
                     BatchNormalization(),
                     ReLU(),
                     Dense(32,kernel_initializer='he_normal',kernel_regularizer=l2(0.001), name='dense3'),
                     BatchNormalization(),
                     ReLU(),
                     Dense(1,kernel_initializer='he_normal', name='output')])
```

```
model4.compile(loss='mse',
              optimizer =Adam(learning_rate=1e-2),
              metrics=['mae'])
```

```
model4.fit(x_train, y_train, epochs=300,
           validation_data=(x_val, y_val))
```

```
model4.evaluate(x_test,y_test)
```

### 모델5
 - dropout을 사용하였으나 성능 저하

```
model5 = Sequential([Dense(64,kernel_initializer='he_normal',kernel_regularizer=l2(0.001), input_shape=(13,), name='dense1'),
                     BatchNormalization(),
                     ReLU(),
                     Dropout(0.5),
                     Dense(64,kernel_initializer='he_normal',kernel_regularizer=l2(0.001), name='dense2'),
                     BatchNormalization(),
                     ReLU(),
                     Dropout(0.5),
                     Dense(32,kernel_initializer='he_normal',kernel_regularizer=l2(0.001), name='dense3'),
                     BatchNormalization(),
                     ReLU(),
                     Dropout(0.5),
                     Dense(1,kernel_initializer='he_normal', name='output')])
```

```
model5.compile(loss='mse',
              optimizer =Adam(learning_rate=1e-2),
              metrics=['mae'])
```

```
model5.fit(x_train, y_train, epochs=300,
           validation_data=(x_val, y_val))
```

```
model5.evaluate(x_test,y_test)
```
